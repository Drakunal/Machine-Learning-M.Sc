```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 

data['mainroad']= label_encoder.fit_transform(data['mainroad']) 
  
```

```python
data['guestroom']= label_encoder.fit_transform(data['guestroom'])
data['basement']= label_encoder.fit_transform(data['basement'])
data['hotwaterheating']= label_encoder.fit_transform(data['hotwaterheating'])
data['airconditioning']= label_encoder.fit_transform(data['airconditioning'])
data['prefarea']= label_encoder.fit_transform(data['prefarea'])
data['furnishingstatus']= label_encoder.fit_transform(data['furnishingstatus'])
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['price','area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']] = scaler.fit_transform(data[['price','area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']])
```

```python
data
```

```python
X = data.drop('price',axis = 1)
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.2, random_state=67)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_trainset, y_trainset)
```

```python
linear_pred = linear_reg.predict(X_testset)
linear_pred[0:5]
```

```python
from sklearn.metrics import r2_score 
r2_score(y_testset, linear_pred)
```
